```python
# Импортируем библиотеки pandas
import pandas as pd

# Загружаем данные из CSV-файла.Используем разделитель '\t' (табуляция), так как колонки в файле разделены табами.
df = pd.read_csv('DB_3.csv', sep='\t')

# Выведем первые 5 строк, чтобы убедиться, что данные загружены корректно
df.head()
```

```python
# Проверка общей информации о датафрейме
df.info()

# Очистка названия колонок (кавычки, начальные и конечные пробелы,символы табуляции). 
# Удаление последней колонки 'Unnamed: 15', которая появилась из-за лишнего разделителя в конце строк
df.columns = df.columns.str.replace('"', '').str.strip().str.replace('\t', ' ')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Проверка названия каолонок после очистки
print("Очищенные названия колонок:")
print(df.columns.tolist())
```

```python
# Вывод первые и последние строки для визуального контроля
print("Первые 5 строк:")
display(df.head())

print("\nПоследние 5 строк:")
display(df.tail())

# Получение основных статистических показатели для числовых признаков
print("\nОписательная статистика:")
df.describe()

# Уникальные значения категориальных признаков
print("\nУникальные значения в 'gender':", df['gender'].unique())
print("Уникальные значения в 'occupation':", df['occupation'].unique())
print("Уникальные значения в 'work_mode':", df['work_mode'].unique())
```

```python
# Удаление строк, где user_id отсутствует
df = df[df['user_id'].notna() & (df['user_id'] != '')].copy()

# Удаление дубликатов по user_id, оставив первую запись
df['user_id'] = df['user_id'].astype(str)
df = df.drop_duplicates(subset='user_id', keep='first').reset_index(drop=True)

# Разделение колонок на категориальные и числовые
categorical_cols = ['gender', 'occupation', 'work_mode']
numerical_cols = [col for col in df.columns if col not in categorical_cols + ['user_id']]

# Пропуски в категориальных признаках заменяем модой
for col in categorical_cols:
    mode_value = df[col].mode()[0] if not df[col].mode().empty else 'Unknown'
    df[col] = df[col].fillna(mode_value)

# Пропуски в числовых признаках заменяем медианой (устойчиво к выбросам)
for col in numerical_cols:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)


# Проверка на оставшиеся пропуски
print("Остаточные пропуски:")
print(df.isnull().sum().sum())

# Вывод основную статистику
print("Описательная статистика после обработки:")
df.describe()
```

```python
import matplotlib.pyplot as plt

# Список числовых колонок для анализа выбросов
numeric_cols_for_outliers = [
    'age',
    'screen_time_hours',
    'sleep_hours',
    'stress_level_0_10',
    'productivity_0_100',
    'exercise_minutes_per_week',
    'social_hours_per_week',
    'mental_wellness_index_0_100'
]

# Настройка графика
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(16, 10))
axes = axes.flatten()

for i, col in enumerate(numeric_cols_for_outliers):
    axes[i].boxplot(df[col], vert=True, patch_artist=True, boxprops=dict(facecolor='lightpink'))
    axes[i].set_title(f'Boxplot: {col}')
    axes[i].set_ylabel('')

plt.tight_layout()
plt.show()
```
